### Height Experiment
- The `height_experiment` function simulates the heights of customers in a theme park using a normal distribution.

#### Type of Random Variable

- **Continuous Random Variable**: Height is a continuous measure, as it can take any value within a given range (e.g., 150.5 cm, 150.55 cm, etc.). It's not confined to fixed, separate values, which makes it continuous rather than discrete. Height naturally varies in a continuous manner; it is not limited to discrete units.

#### Probability Distribution
- **Normal Distribution**: In a large sample of people, the heights are usually distributed in a bell-shaped curve pattern, making the normal distribution a fitting choice for this type of data. The mean and standard deviation can help us define this curve.
> NOTE: The tails of the normal distribution extend indefinitely, theoretically allowing for extremely high or low values that might not be realistic for a measurement like human height. However, the normal distribution is still a good approximation for the heights of people in a large sample.

The Central Limit Theorem supports the notion that the distribution of average heights will tend to be normal, especially if the sample size is large. This is a common distribution for biometric measurements like height.

For a more complete discussion of the normal distribution and its properties, please refer to the [Appendix](#Appendix).

In summary, for the given task of observing the average height of customers in a theme park, a continuous random variable modeled by a normal distribution would be the most appropriate choice. This allows for accurate calculations of the mean and standard deviation, which in turn can offer valuable insights for various applications, such as optimizing the design of theme park attractions for customer safety and comfort.

#### Probabilistic Experiments and Simulation Code

### Alternative Probability Distributions

If we have reason to believe that the distribution of heights is more constrained and doesn't follow the long tails often associated with the normal distribution, we might  consider using a different distribution that better fits the characteristics of our data.

- **Beta Distribution**: If we normalize the heights (say, between 0 and 1 or another specific range), the Beta distribution can model data confined to a finite interval.
- **T-distribution**: It is similar to the normal distribution but has fatter tails. It's often used in smaller sample sizes where the standard deviation is estimated from the data, rather than known beforehand.
- **Truncated Normal Distribution**: This is a normal distribution but with limits on the lower and upper values. This could model the realistic scenario that heights can only fall within a certain range.

Each of these alternatives comes with its own set of parameters and assumptions, so it would be crucial to understand the data's behavior thoroughly before choosing a distribution.

In our specific context of using the data for potential applications like optimizing theme park attractions for customer safety and comfort, a more constrained distribution like the truncated normal or even the Beta (if normalized) might offer a more realistic model of height distribution among theme park attendees.

## A Brief Introduction to the Normal Distribution for Human Height

The height of a group of people is typically modeled using a **Normal Distribution**¹³⁵. This is because human height tends to cluster around a certain average value, while the probability of heights further from the average decreases symmetrically on both sides³.

The Normal Distribution is characterized by two parameters: the **mean** (average) and the **standard deviation**⁴. The mean represents the center of the distribution (the peak), while the standard deviation measures the spread⁴.

For instance, human height is said to be normally distributed with a population mean of 180 cm and standard deviation of 7.62 cm¹. Similarly, the heights of young women aged 20 to 29 follow approximately the N(162.56, 6.858) distribution, and young men the same age have heights distributed as N(175.77, 7.112), where height is measured in centimeters⁵.

It's important to note that these are just examples and the actual parameters would depend on the specific group of people being considered (e.g., their age, gender, ethnicity, etc.). Also, while the Normal Distribution is a common choice due to its mathematical properties and fit to many natural phenomena, it may not perfectly describe all populations or groups⁴. Other distributions could potentially provide a better fit depending on the specific characteristics of the data⁴.

It's valid to use a Normal Distribution because one of the basic rules of probability, known as the **Central Limit Theorem**, states that the distribution of a trait that is determined by independent random variables, like height and genes, roughly follows a bell curve⁶.

The variation of height within a given population is largely determined by genetic factors¹. However, it's important to note that the distribution of human heights is not perfectly normal due to factors such as nutrition and health⁶.

In terms of whether human height should be considered a discrete or continuous random variable, there's some debate. Some argue that because height can be measured with infinite accuracy, it should be considered a continuous random variable¹. However, in practical applications, due to measurement limitations, height is often treated as a discrete variable¹.

Please note that these are general observations and the actual distribution may vary based on the specific population and measurement techniques used.
